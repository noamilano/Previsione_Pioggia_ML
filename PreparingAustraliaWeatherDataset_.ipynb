Questo dataset meteorologico australiano è stato scaricato dal repository Kaggle:
https://www.kaggle.com/datasets/arunavakrchakraborty/australia-weather-data/data.
Contiene circa 10 anni di osservazioni meteorologiche giornaliere provenienti da numerose stazioni meteorologiche in tutta l'Australia.
Le osservazioni giornaliere sono disponibili anche su:
http://www.bom.gov.au/climate/dwo/IDCJDW0000.shtml.

Copyright: Commonwealth of Australia 2010, Bureau of Meteorology.
Licenza del dataset: CC0 – Pubblico Dominio.

Ogni istanza (osservazione) contiene dati meteorologici australiani utilizzati per prevedere la pioggia del giorno successivo.

Descrizione delle variabili:
* Location: Nome della città in Australia.

* MinTemp: Temperatura minima del giorno (°C).

* MaxTemp: Temperatura massima del giorno (°C).

* Rainfall: Quantità di pioggia caduta nel giorno (millimetri).

* Evaporation: Evaporazione totale giornaliera (millimetri).

* Sunshine: Ore di sole registrate in giornata.

* WindGustDir: Direzione della raffica di vento più forte (16 punti cardinali).

* WindGustSpeed: Velocità della raffica più forte (km/h).

* WindDir9am: Direzione del vento nei 10 minuti prima delle 9:00 (punti cardinali).

* WindDir3pm: Direzione del vento nei 10 minuti prima delle 15:00 (punti cardinali).

* WindSpeed9am: Velocità del vento nei 10 minuti prima delle 9:00 (km/h).

* WindSpeed3pm: Velocità del vento nei 10 minuti prima delle 15:00 (km/h).

* Humidity9am: Umidità relativa alle 9:00 (%).

* Humidity3pm: Umidità relativa alle 15:00 (%).

* Pressure9am: Pressione atmosferica alle 9:00 (hPa).

* Pressure3pm: Pressione atmosferica alle 15:00 (hPa).

* Cloud9am: Copertura nuvolosa alle 9:00 (in ottavi).

* Cloud3pm: Copertura nuvolosa alle 15:00 (in ottavi).

* Temp9am: Temperatura alle 9:00 (°C).

* Temp3pm: Temperatura alle 15:00 (°C).

* RainToday: “Yes” se ha piovuto oggi, “No” altrimenti.

* RainTomorrow: 1 se pioverà domani, 0 altrimenti.

Il dataset grezzo originale contiene 99.516 istanze.
È stato ripulito, preprocessato e preparato in questo notebook.

Dopo la fase di preparazione dei dati, si ottiene un dataset finale con 39.574 casi, in cui:

* sono presenti 21 attributi

* ogni attributo è stato normalizzato con Min-Max Scaling:  (x−min)/(max−min)  

Il file risultante viene salvato come: AustraliaWeatherPreparedCleanAttributes.csv

Il problema di classificazione consiste nel prevedere se pioverà domani, con due possibili valori:

* 1 → sì, pioverà

* 0 → no, non pioverà

Queste etichette di output sono memorizzate in:
AustraliaWeatherOutput.csv,
utilizzato per addestrare modelli supervisionati.

```python
# Librerie
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, minmax_scale
import matplotlib.pyplot as plt

# Monta Google Drive
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# Percorsi dei file
INPUT_FILE_NAME = "/content/drive/My Drive/Colab Notebooks/datasets/AustraliaWeather/AustraliaWeatherRawDataset.csv"
ATT_FILE_NAME = "/content/drive/My Drive/Colab Notebooks/datasets/AustraliaWeather/AustraliaWeatherPreparedCleanAttributes.csv"
OUTPUT_FILE_NAME = "/content/drive/My Drive/Colab Notebooks/datasets/AustraliaWeather/AustraliaWeatherOutput.csv"

# Carica il dataset
dataset = pd.read_csv(INPUT_FILE_NAME)

# Esplora i dati
print(dataset.info())
display(dataset.head(10))
display(dataset.tail(10))

# Conta i valori nulli per colonna
nulls = {att: dataset[dataset[att].isnull()].shape[0] for att in dataset.columns}
print("Valori nulli per colonna:\n", nulls)

# Rimuove righe con valori nulli
dataset.dropna(inplace=True)
print("\nDopo la pulizia:")
print(dataset.info())

# Verifica distribuzione target RainTomorrow
print("\nDistribuzione RainTomorrow (dopo pulizia):")
print(dataset["RainTomorrow"].value_counts(normalize=True).round(3))

# Statistiche di frequenza per variabili categoriche
print(dataset.Location.value_counts())
print(dataset.WindGustDir.value_counts())
print(dataset.WindDir9am.value_counts())
print(dataset.WindDir3pm.value_counts())
print(dataset.RainToday.value_counts())
print(dataset.RainTomorrow.value_counts())

# Istogrammi per le principali variabili categoriche
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(10, 10))
fig.tight_layout(h_pad=10, w_pad=5)

ax1.hist(dataset['Location'])
ax1.tick_params(axis='x', rotation=90)
ax1.set_title('Location')

ax2.hist(dataset['WindGustDir'])
ax2.tick_params(axis='x', rotation=90)
ax2.set_title('WindGustDir')

ax3.hist(dataset['WindDir9am'])
ax3.tick_params(axis='x', rotation=90)
ax3.set_title('WindDir9am')

ax4.hist(dataset['WindDir3pm'])
ax4.tick_params(axis='x', rotation=90)
ax4.set_title('WindDir3pm')

ax5.hist(dataset['RainToday'])
ax5.set_title('RainToday')

ax6.hist(dataset['RainTomorrow'])
ax6.set_title('RainTomorrow')

plt.show()

# Rimuove la colonna ID se presente perchè no rilevante
if "row ID" in dataset.columns:
    dataset = dataset.drop("row ID", axis=1)

# Mescola le righe del dataset, per rimuovere eventuali bias di ordinamento
dataset = dataset.sample(frac=1).sample(frac=1).sample(frac=1).reset_index(drop=True)

# Label Encoding per variabili categoriche: gli attributi discreti (categoriali) sono stati codificati con LabelEncoder affinché la rete neurale potesse elaborarne i valori numerici.
encoder_location = LabelEncoder()
encoder_WindGustDir = LabelEncoder()
encoder_WindDir9am = LabelEncoder()
encoder_WindDir3pm = LabelEncoder()
encoder_RainToday = LabelEncoder()

dataset["Location"] = encoder_location.fit_transform(dataset["Location"].values)
dataset["WindGustDir"] = encoder_WindGustDir.fit_transform(dataset["WindGustDir"].values)
dataset["WindDir9am"] = encoder_WindDir9am.fit_transform(dataset["WindDir9am"].values)
dataset["WindDir3pm"] = encoder_WindDir3pm.fit_transform(dataset["WindDir3pm"].values)
dataset["RainToday"] = encoder_RainToday.fit_transform(dataset["RainToday"].values)

# Codifica inversa di esempio
inverse_example = {value: encoder_RainToday.inverse_transform([value])[0] for value in range(2)}
print("Esempio codifica inversa RainToday:", inverse_example)

# Visualizza prime righe
display(dataset.head())

# Heatmap di correlazione
correlation_matrix = dataset.corr()
plt.figure(figsize=(30, 25))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matrice di Correlazione delle Variabili Numeriche')
plt.show()

# Separazione attributi e target, con estrazione del target t dalla colonna RainTomorrow
t = pd.DataFrame(data=dataset["RainTomorrow"].values, columns=["RainTomorrow"])
dataset.drop(columns="RainTomorrow", inplace=True)

# Controllo valori estremi prima della normalizzazione
print("\nStatistiche dei valori numerici prima della normalizzazione:")
display(dataset.describe().loc[["min", "max"]])

# Normalizzazione degli attributi
x = pd.DataFrame(minmax_scale(dataset, feature_range=(-1, 1)), columns=dataset.columns)

# Statistiche descrittive
display(x.describe().round(2))

# Salvataggio su Drive
x.to_csv(ATT_FILE_NAME, index=False)
t.to_csv(OUTPUT_FILE_NAME, index=False)

print("✔️ File salvati correttamente!")
print("→ Attributi:", ATT_FILE_NAME)
print("→ Target:", OUTPUT_FILE_NAME)
```
